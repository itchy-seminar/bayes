```python
import random
import matplotlib.pyplot as plt
```

```python
simple_mean = []
normalized_mean = []
cum_x = 0.0
cum_w = 0.0
for i in range(10 ** 6):
    x = random.random() + random.random() + random.random() + random.random()
    if x <= 1.0:
        cum_x += x * 24.0
        cum_w += 24.0

    if (i + 1) % 10000 == 0:
        simple_mean.append(cum_x / i)
        normalized_mean.append(cum_x / cum_w)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(simple_mean, '+-', label='simple')
ax.plot(normalized_mean, 'x-', label='normalized')
ax.legend()
ax.grid()
```
