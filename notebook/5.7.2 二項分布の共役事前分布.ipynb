```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

# 問題設定

コイン投げでコインの表が出る確率 $p$ を求めたい。

```python
# p の(共役)事前分布
def pi(p, alpha: float, beta: float):
    return p ** (alpha - 1) * (1 - p) ** (beta - 1) / scipy.special.beta(alpha, beta)
```

```python
p = np.arange(0, 1, 0.01)
plt.plot(p, pi(p, 1, 1))
```

```python
# 10回降って9回表が出た場合の事後分布
plt.plot(p, pi(p, 1 + 9, 1 + 1))
plt.grid()
```

```python
np.argmax(pi(p, 1 + 9, 1 + 1))
```

```python
a0 = b0 = 1
# 10回降って9回表が出た
a1 = a0 + 9
b1 = b0 + 10 - 9

# 新しく10回振る時の事後分布
n = 10
x = np.arange(0, n + 1)
y = scipy.special.comb(n, x) * scipy.special.beta(a1 + x, b1 + n - x) / scipy.special.beta(a1, b1)
plt.scatter(x, y)
plt.grid()
```

```python
# 6回表が出た
a2 = a1 + 6
b2 = b1 + n - 6
```

```python
plt.plot(p, pi(p, a1, b1), label="1")
plt.plot(p, pi(p, a2, b2), label="2")
plt.legend()
plt.grid()
```

```python
n = 10
x = np.arange(0, n + 1)
y = scipy.special.comb(n, x) * scipy.special.beta(a1 + x, b1 + n - x) / scipy.special.beta(a1, b1)
plt.scatter(x, y, label="1")
y = scipy.special.comb(n, x) * scipy.special.beta(a2 + x, b2 + n - x) / scipy.special.beta(a2, b2)
plt.scatter(x, y, label="2")
plt.legend()
plt.grid()
```
